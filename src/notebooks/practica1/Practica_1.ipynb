## Práctica 1: Descripción y estructura

### Definición de URDF

El **Formato de Descripción Robótica Unificada** (URDF, por sus siglas en inglés) es un archivo en **lenguaje de marcado extensible (XML)** que incluye la descripción física de un robot. Es un modelo 3D que contiene información sobre articulaciones, motores, masa, entre otros. Estos archivos son ejecutados a través del **Sistema Operativo del Robot (ROS)**, proporcionando información clave al operador sobre la estructura y capacidades del robot antes de ser operado.

### Tipos de juntas definidas en URDF

- **Revolute**: Junta rotacional que permite el giro sobre un eje con un rango limitado especificado por los límites superior e inferior.
- **Continuous**: Junta rotacional similar a la **revolute**, pero sin límites superior e inferior.
- **Prismatic**: Junta que permite el deslizamiento a lo largo de un eje con un rango limitado por los límites.
- **Fixed**: Junta fija que no permite movimiento, útil para asociar sistemas de referencia, como sensores o cámaras.
- **Floating**: Junta que permite el movimiento en 6 grados de libertad.
- **Planar**: Junta que permite el movimiento en un plano perpendicular al eje.

### Qué es un nodo y qué es un tópico en ROS 2

- **Nodo**: Los nodos son unidades de cómputo dentro del sistema ROS. Publican en **tópicos** para enviar datos o se suscriben a estos para recibir información. También pueden actuar como clientes o servidores de servicios, realizando cálculos u ofreciendo funcionalidad a otros nodos.
  
- **Tópico**: Es un sistema de **publicación/suscripción** donde los productores de datos (publicadores) y los consumidores de datos (suscriptores) se comunican a través de un tema común. Todos los publicadores y suscriptores que comparten el mismo nombre de tema pueden comunicarse directamente entre sí. Puede haber cero o más publicadores y cero o más suscriptores en un tema en particular. Cuando cualquiera de los publicadores publica datos en el tema, todos los suscriptores del sistema recibirán los datos. Los publicadores y suscriptores pueden ir y venir según sea necesario, lo que significa que la depuración y la introspección son extensiones naturales del sistema.

### Definición de RVIZ

**RVIZ** es una herramienta de visualización 3D para ROS. Proporciona una vista del modelo simulado de cualquier robot descrito por URDF, y también permite visualizar datos de sensores. Esta herramienta es crucial para la depuración de aplicaciones robóticas, ya que permite ver lo que el robot está "pensando" y "haciendo" en tiempo real. Tiene la particularidad de percibir los errores ocasionados por el modelo del robot o sensores en tiempo real.
